# 最值归一化

```python
(x - np.min(x))/(np.max(x) - np.min(x))
```

```python
X = np.random.randint(0, 100, size=(50, 2))
X[:10, :]
```

```python
# 由于生成的数组是整型，归一化后是浮点型，所以需要对X进行一个强制类型转换
X = np.array(X, dtype="float")
X[:10, :]
```

```python
x = np.random.randint(0, 100, size=(50, 2))
x = np.array(x, dtype="float")
x[:10, :]
```

```python
for i in range(x.shape[1]):
    x[:, i] = (x[:, i] - np.min(x[:, i]))/(np.max(x[:, i]) - np.min(x[:, i]))
```

```python
x[:10, :]
```

```python
plt.scatter(x[:, 0], x[:, 1])
```

# 均值方差归一化 Standardization

```python
Y = np.random.randint(0, 100, size=(50, 2))
Y = np.array(Y, dtype="float")
Y[:10, :]
```

```python
for i in range(Y.shape[1]):
    Y[:, i] = (Y[:, i] - np.mean(Y[:, i]))/np.std(Y[:, i])
```

```python
Y[:10, :]
```

```python
plt.scatter(Y[:, 0], Y[:, 1])
```

# 使用sklearn中的封装的归一化包

PS：有个主意的点，我们计算出训练数据集的平均值和方差要要进行保存，因为他们要使用到测试数据集中去做归一化，就是说测试集的归一化是使用训练集的均值和方差

```python
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
x1 = iris.data
y1 = iris.target
```

```python
x1[:10, :]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.2, random_state=666)  # random_state=666就是随机种子
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
standardScaler.fit(X_train)
```

```python
standardScaler.mean_  # 均值
```

```python
standardScaler.scale_  # 方差
```

```python
standardScaler.transform(X_train)  # 对训练数据集归一化处理
```

```python
X_train  # 说明未对原始数据进行修改
```

```python
X_train_standard = standardScaler.transform(X_train)
X_test_standard = standardScaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=3)
knn_clf.fit(X_train_standard, y_train)
y_predict = knn_clf.predict(X_test_standard)
sum(y_predict == y_test)/len(y_test)
```
